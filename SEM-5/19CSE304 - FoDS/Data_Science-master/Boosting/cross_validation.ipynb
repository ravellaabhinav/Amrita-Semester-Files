# Leave-P-Out Cross Validation

```
from sklearn.model_selection import LeavePOut
import numpy as np
import pandas

from google.colab import drive
drive.mount('/content/drive')


# Importing sample data from a csv to DataFrame
DataFrame = pandas.read_csv('/content/drive/My Drive/Boosting/Random Sample.csv')
DataFrame=DataFrame.head(5) #Taking 5 samples for better understanding

# creating design matrix X and target vector y

X = np.array(DataFrame.iloc[:, 0:4]) # end index is exclusive
Y = np.array(DataFrame['House']) # taking House as target

leaveout = LeavePOut(3) # taking p=2
leaveout.get_n_splits(X) # Number of splits of X

# Printing the Train & Test Indices of splits
for train_index, test_index in leaveout.split(X): 
   print("TRAIN:", train_index, "TEST:", test_index)
   X_train, X_test = X[train_index], X[test_index]
   Y_train, Y_test = Y[train_index], Y[test_index]
```

# K-Fold Cross Validation
